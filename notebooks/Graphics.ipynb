```python
import re

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
d90_file = '../results/complexity_utf-16_1010_10_10_bibles_90_lcm.csv'
dall_file = '../results/complexity_utf-16_2020_10_10_bibles_lcm.csv'

df90 = pd.read_csv(d90_file, index_col=False)
df90 = df90[df90.language != 'NAMBIKUÁRA'] # We removed Nambikuára because it has tone annotation
df90.loc[df90.metric == 'del-chars', 'value'] = - df90.loc[df90.metric == 'del-chars', 'value']

dfall = pd.read_csv(dall_file, index_col=False)
dfall = dfall[dfall.language != 'NAMBIKUÁRA'] # We removed Nambikuára because it has tone annotation
dfall.loc[dfall.metric == 'del-chars', 'value'] = - dfall.loc[dfall.metric == 'del-chars', 'value']

family = pd.read_csv('../dataset/family.csv', index_col=None)
```

```python
def compute(dfi, alg):
    df = dfi.copy()
    ag = df.algorithm == alg
    dw = df.metric == 'del-words'
    dc = df.metric == 'del-chars'
    
    x = df[dw & ag].groupby(['language', 'algorithm']).value.transform('mean').to_numpy()
    y = df[dc & ag].groupby(['language', 'algorithm']).value.transform('mean').to_numpy()

    df.loc[dw & ag, 'del-words'] = x
    df.loc[dw & ag, 'del-chars'] = y
    df = df.dropna().drop_duplicates('language').reset_index().drop(columns=['metric', 'index', 'run_id'])

    return df
```

```python
def plot(df90, dfall, algo, family, wals=None, text=False):

    def process_df(df, algo, wals):
        wals = wals or set(df.wals)
        df = compute(df[df.wals.isin(wals)], algo)
        df = df.merge(family[['wals', 'family']], on='wals', how='left')
        return df

    def create_color_marker_dicts(df1, df2):
        unique_families = list(set(list(df1['family'].unique()) + list(df2['family'].unique())))
        colors = plt.cm.tab20(np.linspace(0.0, 1, len(unique_families)))
        markers = ['o', 'p', 'd', 'P'] + [(4, 1, 0), (6, 1, 0)] + [(i, 2, 0) for i in range(3, 5)] + [(i, 2, 0) for i in range(6, 8)] +  [(i, 0, 0) for i in range(3, 7)] + ['x', 'v', 'X', 8, 9]#for j in range(0, 2)][::-1]
        family_to_color = { f: colors[i] for i, f in enumerate(unique_families)}
        family_to_marker = { f : markers[i] for i, f in enumerate(unique_families)}
        return unique_families, (colors, family_to_color), (markers, family_to_marker)

    def set_color_marker(df, family_to_color, family_to_marker): 
        df['color'] = df['family'].map(family_to_color)
        df['marker'] = df['family'].map(family_to_marker)
        return df

    # process
    df90 = process_df(df90, algo, wals)
    dfall = process_df(dfall, algo, wals)

    unique_families, (colors, f2c), (markers, f2m) = create_color_marker_dicts(df90, dfall)

    df90 = set_color_marker(df90, f2c, f2m)
    dfall = set_color_marker(dfall, f2c, f2m)
    
    
    fig, ax = plt.subplots(2, 1, figsize=(15, 8.0), sharex=False, sharey=True)
    fig.suptitle(r"Trade-off hypothesis ($\mathcal{H}_3$) with $\it{%s}$" % (algo.lower(),), fontsize=30)
    titles = ['D90 subset', 'DALL subset']
    for di, df in enumerate([df90, dfall]):
        x = df['del-chars'].to_numpy()
        y = df['del-words'].to_numpy()
        l = df['wals'].to_numpy()
        # c = df['color'].to_numpy()
        ax[di].set_xlim((-1.3, -1.05))
        for i, f in enumerate(unique_families):
            df_family = df[df['family'] == f]
            if len(df_family):
                ax[di].scatter(
                    df_family['del-chars'],
                    df_family['del-words'],
                    c=[colors[i]],
                    marker=markers[i],
                    label=f if di == 1 else "",
                    s=150,
                )
        if di == 1: ax[di].set_xlabel(r'Morphological Complexity $\mu^{\mathbb{M}}_{\times}$', weight='bold', fontsize=18)
        ax[di].set_title(titles[di], fontsize=16, weight='bold')
        ax[di].set_ylabel(r'Syntatic Complexity $\mu^{\mathbb{S}}_{\times}$', weight='bold', fontsize=18)

        if di == 0:
            delta = 1e-3
            rot = 0
            dec = lambda u, f: u - f*delta
            inc = lambda u, f: u + f*delta
            for xi, yi, li in zip(x, y, l):
                match li :
                    case "[gnw]":
                        xi = dec(xi, 8)
                        yi = dec(yi, 5)
                    case "crg":
                        xi = inc(xi, 1)
                        yi = inc(yi, 1)
                    case "fre":
                        xi = dec(xi, 8.0)
                        yi = dec(yi, 1.0)
                    case "spa":
                        xi = dec(xi, 9)
                        yi = dec(yi, 2)
                    case "ger":
                        xi = inc(xi, 1)
                        yi = dec(yi, 4)
                    case "[yrl]":
                        yi = dec(yi, 2)
                        xi = dec(xi, 11)
                    case "apu":
                        xi = dec(xi, 11)
                        yi = dec(yi, 2)
                    case "[grc]":
                        xi = dec(xi, 4)
                        yi = dec(yi, 5)
                    case "apl":
                        xi = inc(xi, 2)
                    case "plk":
                        xi = inc(xi, 1.5)
                    case "jmm":
                        xi = dec(xi, 4)
                        yi = inc(yi, 2.5)
                    case "pex":
                        yi = dec(yi, 6)
                        xi = dec(xi, 3)
                    case "wps":
                        yi = dec(yi, 4)
                        xi = inc(xi, 1.5)
                    case "[guu]":
                        xi = inc(xi, 2)
                        yi = dec(yi, 0.5)
                    case "kng":
                        xi = inc(xi, 2)
                        yi = dec(yi, 0.5)
                    case "mac":
                        xi = dec(xi, 11)
                        yi = dec(yi, 2)
                    case _:
                        xi = inc(xi, 2)
                        yi = dec(yi, 1)
                ax[di].text(xi, yi, li, rotation=rot, fontsize=16, weight='bold')
    legend = fig.legend(title="Language Families", fontsize=30, bbox_to_anchor=(0.4, 0.3, 0.5805, 0.5905), ncol=3, prop={'weight' : 'bold'})
    #legend = fig.legend(fontsize=12, bbox_to_anchor=(0.4, 0.3, 0.58, 0.600), ncol=3, prop={'weight' : 'bold'})
    legend.get_frame().set_alpha(None)
    fig.tight_layout()
    fig.savefig("../images/h3.pdf")
```

```python
plot(df90, dfall, 'gzip', family=family, text=True)
```
